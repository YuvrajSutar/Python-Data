# Data Structure Overview:
    

- 1-List: It is similar to array the data elements can be of different data types.
    
- 2-Tuple: similar to lists but they are immutable which means the values in a tuple cannot be modified they can only be read.
    
- 3-Dictionary: The dictionary contains Key-value pairs as its data elements.


# Key Points to Remember

- Mutable-can change or update after declaration e.g. list, mutable set, dictionary
- Immutable-can't change or update after declaration e.g. number,string,tuple
- Ordered Sequences-Indexes starting from 0 e.g. string,list,tuple
- unordered dequences-Indexes does't start with 0 e.g set & dictionary
- Hashable-Can access values from their indices
- Unhashable-Can't access values from their indices e.g. Numbers,sets
